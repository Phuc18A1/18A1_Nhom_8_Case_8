# Đọc các file

```python
import pandas as pd

product_df = pd.read_csv("product_catalog_clean.csv")
order_df   = pd.read_csv("order_details_clean.csv")
refund_df  = pd.read_csv("return_refund_clean.csv")
```

# 1.Merge product_catalog với order_details.

```python
order_product = order_df.merge(
    product_df,
    on="product_id",
    how="left"
)
```

# 2.Merge thêm return_refund theo order_id & product_id.

```python
order_refund = order_product.merge(
    refund_df,
    on="product_id",
    how="left"
)
```

# 3.Phát hiện mismatch giữa đơn đặt và đơn trả hàng

```python
order_refund["is_returned"] = order_refund["refund_amount"].notna()

order_refund["mismatch"] = (
    order_refund["is_returned"] &
    (order_refund["quantity"] > 0)
)
order_refund["revenue"] = (
    order_refund["quantity"] * order_refund["unit_price"]
)
```

# Xuất file hoàn chỉnh.

```python
order_refund.to_csv(
    "final_order_report.csv",
    index=False
)

print("Xuất file final_order_report.csv thành công")
```
