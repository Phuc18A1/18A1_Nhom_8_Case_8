# Đọc các file

```python
import pandas as pd
order_details = pd.read_csv("order_details_clean.csv")
product_catalog = pd.read_csv("product_catalog_clean.csv")
return_refund = pd.read_csv("return_refund_clean.csv")
```

# 1.Merge product_catalog với order_details.

```python
order_product = order_details.merge(
    product_catalog,
    on="product_id",
    how="left"
)
```

# 2.Merge thêm return_refund theo order_id & product_id.

```python
order_product_return = order_product.merge(
    return_refund,
    on=["order_id", "product_id"],
    how="left",
    suffixes=("", "_return"),
    indicator="return_merge"
)
```

# 3.Phát hiện mismatch giữa đơn đặt và đơn trả hàng

```python
mismatch_returns = return_refund.merge(
    order_details,
    on=["order_id", "product_id"],
    how="left",
    indicator=True
).query("_merge == 'left_only'")

# Gắn cờ mismatch trong bảng chính
order_product_return["return_mismatch"] = (
    order_product_return["return_merge"] == "right_only"
)
```

# Xuất file hoàn chỉnh.

```python
order_product_return.to_csv(
    "final_report.csv",
    index=False,
    encoding="utf-8-sig"
)

print(" Đã merge & xuất file order_product_return_full.csv")
print("Số dòng mismatch:", mismatch_returns.shape[0])
```
